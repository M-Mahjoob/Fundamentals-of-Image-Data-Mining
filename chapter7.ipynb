```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.mixture import GaussianMixture
```

```python
# Define means and covariances
mu = np.array([[1, 2], [-3, -5]])
sigma = np.array([[[2, 0.5], [0.5, 1]], [[1, 1], [1, 1]]])

# Create Gaussian Mixture Model
gmm = GaussianMixture(n_components=3, means_init=mu, covariance_type='full', random_state=42)

# Generate data points for visualization
x, y = np.meshgrid(np.linspace(-10, 10, 100), np.linspace(-10, 10, 100))
xy = np.column_stack([x.flatten(), y.flatten()])

# Fit the model
gmm.fit(xy)

# Compute the probability density function (PDF) values
pdf_values = np.exp(gmm.score_samples(xy))
pdf_values = pdf_values.reshape(x.shape)

# Plot the 3D surface
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, pdf_values, cmap='viridis', alpha=0.8)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('PDF Value')
ax.set_title('Gaussian Mixture Model in 3D')

# Show the plot
plt.show()

# Visualize the model from different angles
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

for angle in range(0, 360, 10):
    ax.view_init(30, angle)
    ax.plot_surface(x, y, pdf_values, cmap='viridis', alpha=0.8)
    plt.draw()
    plt.pause(0.01)

# Show the 2D plot
plt.figure(figsize=(8, 8))
plt.contourf(x, y, pdf_values, cmap='viridis')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Gaussian Mixture Model in 2D')

# Show the plot
plt.show()
```
