```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img_1 = cv2.imread("C4.1_img.png",0)
img_2 = cv2.imread("C4.2_img.png",0)
img_3 = cv2.imread("C4.3_img.png",0)
```

```python
plt.subplot(3,4,1)
plt.imshow(img_1,cmap='gray')
plt.title('image_1')
plt.xticks([])
plt.yticks([])

plt.subplot(3,4,2)
plt.hist(img_1.ravel(),256,[0,255])
plt.title('histogram\n')

plt.subplot(3,4,3)
plt.hist(img_1.ravel(),256,[0,255],(0,1))
plt.title('normalized\n histogram')

plt.subplot(3,4,4)
plt.hist(img_1.ravel(),256,[0,255])
plt.xscale('log')
plt.title('logarithm\n histogram')

plt.subplot(3,4,5)
plt.imshow(img_2,cmap='gray')
plt.title('image_2')
plt.xticks([])
plt.yticks([])

plt.subplot(3,4,6)
plt.hist(img_2.ravel(),256,[0,255])

plt.subplot(3,4,7)
plt.hist(img_2.ravel(),256,[0,255],(0,1))

plt.subplot(3,4,8)
plt.hist(img_2.ravel(),256,[0,255])
plt.xscale('log')

plt.subplot(3,4,9)
plt.imshow(img_3,cmap='gray')
plt.title('image_3')
plt.xticks([])
plt.yticks([])

plt.subplot(3,4,10)
plt.hist(img_3.ravel(),256,[0,255])

plt.subplot(3,4,11)
plt.hist(img_3.ravel(),256,[0,255],(0,1))

plt.subplot(3,4,12)
plt.hist(img_3.ravel(),256,[0,255])
plt.xscale('log')

plt.tight_layout()
plt.show()
```
